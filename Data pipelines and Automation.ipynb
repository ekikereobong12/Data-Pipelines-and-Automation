```python
import requests
import pandas as pd

# Your API key has been inserted here
API_KEY = " 
BASE_URL = "http://api.openweathermap.org/data/2.5/weather"

# List of at least 3 cities
cities = ["Lagos", "Nairobi", "London"]
weather_data = []

for city in cities:
    params = {
        "q": city,
        "appid": API_KEY,
        "units": "metric"  # This gives you temperature in Celsius
    }
    
    response = requests.get(BASE_URL, params=params)
    
    if response.status_code == 200:
        data = response.json()
        # Extract the fields you need
        weather_data.append({
            "City Name": data["name"],
            "Temperature": data["main"]["temp"],
            "Humidity": data["main"]["humidity"],
            "Weather Condition": data["weather"][0]["description"],
            "Wind Speed": data["wind"]["speed"],
            "Date and Time": data["dt"]
        })
    else:
        print(f"Could not retrieve data for {city}")

# Create a Pandas DataFrame for transformation
df = pd.DataFrame(weather_data)
print(df)

# To save your data (Task 3)
# df.to_csv("weather_data.csv", index=False)
```

```python
# Convert Unix timestamp to readable date and time
df['Date and Time'] = pd.to_datetime(df['Date and Time'], unit='s')

# Display the cleaned DataFrame
print("Transformed Data:")
print(df)

# Now you can uncomment this line to save your clean data (Task 3)
# df.to_csv("weather_data.csv", index=False)
```

```python
df.to_csv("weather_data.csv", index=False)
```

```python
# Find the city with the highest humidity
highest_humidity_city = df.loc[df['Humidity'].idxmax()]
print("City with the highest humidity:")
print(highest_humidity_city[['City Name', 'Humidity']])
```
